```python
#1-Load dataset, Show first 5 rows

import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
#4-Data Cleaning
```

```python
#Filling missing age with median 
# Syntex
        #df['column'].fillna(value, inplace=True)

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode())

df.isnull().sum()
```

```python
## Drop Cabin (too many missing values) (not useful for analysis)
#Syntex
        df.drop(columns=['column1', 'column2'], inplace=True)

df.drop(columns=['Cabin'], inplace=True)
print(df.columns)
```

```python
## Convert Sex to numeric
# Syntex
        # df['column'] = df['column'].map({'value1': 0, 'value2': 1})

df['Sex']=df['Sex'].map({'female':1,'male':0})
df.info()
```

```python
#5- Basic Data Analysis
# Syntex
     # df.groupby('column')['target'].mean()

# Survival rate by gender
print(df.groupby('Sex')['Survived'].mean())
```

```python
#6- Visualization
# install pyplot
        # pip install matplotlib

# Install seaborn, Built on top of Matplotlib, but much easier and better for statistical visualizations.  
        #>> pip3 install seaborn

```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Survival count
# Syntex
        # sns.countplot(data=df, x='column', hue='target')

sns.countplot(data=df, x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
# Survival by Sex
sns.countplot(data=df,x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.show()
```

```python
# Age distribution by Survived
# Histogram syntex
    # sns.histplot(data=df, x='column', hue='target', kde=True)

sns.histplot(data=df,x='Age',hue='Survived', kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Feature Correlation
#Heatmap syntex
        # sns.heatmap(df.corr(), annot=True)
```

```python
# Step 1: Check the actual data types
print(df.dtypes)

# Step 2: Force selection of numeric columns safely
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Step 3: Display the numeric columns for verification
print("Numeric Columns Selected:\n", numeric_df.columns)

# Step 4: Generate the heatmap
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```
